# 02 — Descriptive analysis
Do not start advanced models until these distributions and missingness patterns make sense.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../data/processed/core_analytic_dataset.csv')
df.groupby('screen_positive_bp')[['age_years','bmi_kg_m2','waist_cm','mean_sbp','mean_dbp']].agg(['count','mean','std','median'])
```

```python
plt.figure(figsize=(7,4))
plt.hist(df['mean_sbp'].dropna(), bins=25)
plt.xlabel('Mean systolic BP (mm Hg)')
plt.ylabel('Participants')
plt.title('Distribution of Mean Systolic Blood Pressure')
plt.show()
```

```python
missing = df.isna().mean().sort_values(ascending=False)
missing[missing > 0].to_frame('missing_fraction')
```
